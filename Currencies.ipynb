```python
import requests
import xmltodict

url = "http://www.cbr.ru/scripts/XML_daily.asp"
response = requests.get(url)
response
```

```python
response.content
```

```python
data = xmltodict.parse(response.content)
data
```

```python
items = []
for item in data['ValCurs']['Valute']:
    items.append(item)
items
```

```python
my_array = []
for item in data['ValCurs']['Valute']:
    my_set = [item['CharCode'], item['Name'], item['Value']];
    my_array.append(my_set)
    print(my_set)
```
